# 1) 의미 동일성 필터링

```python
# api 호출 함수 정의

from openai import OpenAI

client = OpenAI(
    # defaults to os.environ.get("OPENAI_API_KEY")
    api_key="",
)

def gpt_response(text1, text2):
    prompt = f"text1 = \"{text1}\" \n text2 = \"{text2}\""

    # OpenAI API 호출
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "영어로 작성된 text1과 영어와 한국어가 섞인 text2를 줄게. 두 텍스트의 의미가 동일한지 확인해줘. 동일하면 'Y', 동일하지 않으면 'N'로 답해줘."},
            {"role": "user", "content": prompt}
        ]
    )
    answer = response.choices[0].message.content
    return answer
```

```python
print(gpt_response(df['note'][0], df['mixed'][0]))
```

```python
df["semantic"] = ""

for i in range(len(df)):
    df["semantic"][i] = gpt_response(df['note'][i], df['mixed'][i])
```

```python
print((df['semantic_1'] == 'N').sum())
```

```python
df["semantic_1"] = ""

for i in range(len(df)):
    df["semantic_1"][i] = gpt_response(df['note'][i], df['mixed'][i])

df.to_excel('gpt_data_.xlsx')
```

```python
df['len'] = 0
for i in range(len(df)):
    print(i)
    df['len'][i] = len(str(df['mixed'][i]).split())
```

```python
import re

# 정규표현식 패턴
pattern = "([a-zA-Z]\ ?[ㄱ-ㅎ|ㅏ-ㅣ|가-힣])|([ㄱ-ㅎ|ㅏ-ㅣ|가-힣] ?[a-zA-Z])"
```

```python
df['shift_cnt'] = 0
for i in range(len(df)):
    print(i)
    matches = re.findall(pattern, str(df['mixed'][i]))
    df['shift_cnt'][i] = len(matches)
```

```python
df['shift_rate'] = df['shift_cnt']/df['len']
```
